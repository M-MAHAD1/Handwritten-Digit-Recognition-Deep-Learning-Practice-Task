```python
from keras.models import load_model
from keras.utils import load_img
from keras.utils import img_to_array
import numpy as np
```

```python
# 1. Load the trained model for prediction 
```

```python
model = load_model('HandDigitRecognition.h5')
model.summary()
```

```python
img = load_img('digit0.jpeg', color_mode='grayscale')
img
```

```python
NewImage = img_to_array(img)
NewImage.shape
```

```python
NewImage = NewImage.reshape((1,28*28))
NewImage = NewImage.astype('float32') / 255
```

```python
NewImage.shape
```

```python
# Prediction

prediction = model.predict(NewImage)
prediction
```

```python
NameOfClasses = ['Zero','One','Two','Three','Four',
                 'Five','Six','Seven','Eight','Nine']

digit = NameOfClasses[np.argmax(prediction)]
digit
```

```python
img
```

```python
print(f'Digit is:{digit}')
```
